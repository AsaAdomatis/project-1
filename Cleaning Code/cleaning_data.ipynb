```python
# adding awards as individual columns
# PB: Proball
# AP: All Pro 
# OroY: Offensive rookie of the year
# OPoY: Offensive Player of the year
# CPoY: Comeback player of the year

filtered_df['Awards'] = filtered_df.apply(lambda x: str(x['Awards']).replace(' ', ','), axis=1)

# getting a list of all possible awards given
awards = filtered_df.loc[filtered_df['Awards'].notna()]['Awards']
possible_awards = []
for award_list in awards:
    award_given_list = award_list.split(sep=',')
    for award_given in award_given_list:
        if '-' in award_given:
            award_given = award_given.split(sep='-')[0]
            award_given = award_given.strip()
        if award_given not in possible_awards:
            possible_awards.append(award_given)

possible_awards
```

```python
# applying award ammounts
def get_award_count(row, award_name):
    # Catch Empty
    if row['Awards'] is None:
        return 0

    # Check for specific award
    index = row['Awards'].find(award_name)
    if index >= 0:
        # getting section containing named award at the beginning
        award = row['Awards'][index:]

        # removing multiple entries after award if they exist
        if ',' in award:
            award = award.split(sep=',')[0]

        # if multiple times awarded, get the amount
        if '-' in award:
            return int(award.split(sep='-')[1])
        else:
            return 1
        
    # if no awards by that name, return 0
    else:
        return 0
    
possible_awards.remove('nan')
possible_awards
```

```python
# testing to make sure it works
display(filtered_df.iloc[0, :])
display(get_award_count(filtered_df.iloc[0, :], 'AP'))
display(get_award_count(filtered_df.iloc[0, :], 'PB'))
```

```python
# adding in new awards
for award in possible_awards:
    filtered_df[award] = filtered_df.apply(lambda row: get_award_count(row, award), axis=1)

# checking for correct entry
filtered_df.head()
```

```python
# calculating total awards
def total_rows(row):
    total = 0
    for award in possible_awards:
        total += row[award]
    return total

filtered_df['Total Awards'] = filtered_df.apply(lambda row: total_rows(row), axis=1)
```

```python
# dropping the original awards column
filtered_df = filtered_df.drop(columns='Awards')

filtered_df.head(10)
```

```python
filtered_df.to_csv('Runningback Salary Cap 2017-2018 w Awards.csv')
```
